# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
urunsayisi=int(input("kaç ürün aldınız?"))
with open("urunler.txt","w",encoding="utf-8") as file:
    for adet in range(urunsayisi):
        urun=input("ürün adı nedir")
        fiyat=input("ürünün fiyatı nedir")
        file.write(f"{urun}\n{fiyat}\n")

with open("urunler.txt","r",encoding="utf-8") as file:
    print(file.read())
    
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
    def __str__(self):
        return f"Name: {self.name} Speed: {self.speed} Mileage: {self.mileage}"
    
class Bus(Vehicle):
    def __init__(self, name, speed, mileage,capacity):
        super().__init__(name, speed, mileage)
        self.capacity=capacity

arac=Bus("school volvo",180,12,50)
print(arac)
````
